```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path

import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as st
from scipy.stats import linregress
import seaborn as sns
import hvplot.pandas
import geoviews as gv




# File to Load (Remember to Change These)
all_perth_house = Path("resources/all_perth_310121.csv")

# Read all perth houses sales prices and store into Pandas DataFrames
house_prices_perth = pd.read_csv(all_perth_house)


```

```python
# Specify the columns for display
columns_to_display = ["ADDRESS", "LATITUDE", "LONGITUDE", "SUBURB", "BEDROOMS",
                      "BATHROOMS", "GARAGE", "LAND AREA", "FLOOR AREA", "BUILD YEAR", "KM FROM CBD",
                      "CBD DIST", "DATE SOLD", "POSTCODE", "NEAREST SCH", "NEAREST SCH RANK",
                      "NEAREST SCH DIST"]

# Remove underscores from column names
house_prices_perth.columns = [col.replace("_", " ") for col in house_prices_perth.columns]

# clean the cell Data sold 
 
house_prices_perth ['DATE SOLD']= house_prices_perth ['DATE SOLD'].str.replace ('\r','')

house_prices_perth ['BUILD YEAR']= house_prices_perth ['BUILD YEAR'].fillna(0).astype(int)

#house_prices_perth ["PRICE"] = house_prices_perth  ["PRICE"].map ("${:,.2f}".format)
 

# Convert 'CBD DIST' from meters to kilometers
house_prices_perth['KM FROM CBD'] = house_prices_perth['CBD DIST'] / 1000

# Fill NaN values in 'GARAGE' column with 0
house_prices_perth['GARAGE'].fillna(0, inplace=True)

# Display the DataFrame with selected columns
selected_columns_df = house_prices_perth[columns_to_display]

house_prices_perth_cleaned = house_prices_perth.dropna(subset=['NEAREST SCH RANK', 'BUILD YEAR'])
#house_prices_perth_cleaned
```

```python
# Define the start and end dates of the range you want to select.
start_date = '2010-01-01'
end_date = '2020-12-31'

house_prices_perth_cleaned['DATE SOLD'] = pd.to_datetime(house_prices_perth_cleaned['DATE SOLD'])

# Filtra el DataFrame original usando loc
df_cleaned = house_prices_perth_cleaned.loc[
    (house_prices_perth_cleaned['DATE SOLD'] >= start_date) &
    (house_prices_perth_cleaned['DATE SOLD'] <= end_date)
].copy() 

df_cleaned
```

```python
#Mean prices  < 10km  radius from Perth

median_prices_bedroom_df =  df_cleaned[(df_cleaned['KM FROM CBD'] <=10 )]
summary10km_mean_df = median_prices_bedroom_df.groupby(['SUBURB']).agg({
    
    'BATHROOMS': 'mean',
    'BEDROOMS': 'mean',
    'LAND AREA': 'mean',
    'LATITUDE': 'mean',
    'LONGITUDE': 'mean',
    'PRICE': 'mean'
}).reset_index()


summary10km_mean_df.rename(columns={
    'LAND AREA': 'MEAN_LAND_AREA',
    'LATITUDE': 'MEAN_LATITUDE',
    'LONGITUDE': 'MEAN_LONGITUDE',
    'PRICE': 'MEAN_PRICE'
}, inplace=True)



summary10km_mean_df
```

```python
# Sort and display the top 10 high price houses.


Top10_suburb  = summary10km_mean_df.nlargest(10,'MEAN_PRICE')

summary10km_mean_df  ['MEAN_PRICE'] = summary10km_mean_df ['MEAN_PRICE'].map ("${:,.2f}".format)

Top10_suburb
```

```python

# Sort and display the top 10 low price houses.

Top10_lowest_suburb = summary10km_mean_df.nsmallest(10, 'MEAN_PRICE')
Top10_lowest_suburb
```

```python



fig, ax1 = plt.subplots(figsize=(10, 6))

# Grafic The 10 Most Expensive Suburbs
ax1.barh(Top10_suburb['SUBURB'], Top10_suburb['MEAN_PRICE'], color='lightblue')
ax1.set_xlabel('Mean Price')
ax1.set_title('The 10 Most Expensive Suburbs -10 km')
ax1.invert_yaxis()  
plt.gca().xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: '${:,.0f}'.format(x)))


fig, ax2 = plt.subplots(figsize=(10, 6))

# Grafic The 10 Most Lowest Suburbs
ax2.barh(Top10_lowest_suburb['SUBURB'], Top10_lowest_suburb['MEAN_PRICE'], color='blue')
ax2.set_xlabel('Mean Price')
ax2.set_title('The 10 Suburbs with the Lowest Prices - 10 km')
ax2.invert_yaxis()  
plt.gca().xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: '${:,.0f}'.format(x)))

# show a grafic
plt.tight_layout()
plt.show()
```
